```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('country details for pyton program.xlsx', header = 4,
                    usecols = ['COUNTRY','INFECTED','TESTED','RECOVERED','DECEASED'])
df
```

```python
df.info()
```

```python
df["INFECTED"][12]=np.nan
df["TESTED"][10]=np.nan
```

```python
df=df.fillna(0)
df
```

```python
df.info()
```

```python
df["RECOVERED"] = df.RECOVERED.astype(int)
df["DECEASED"] = df.DECEASED.astype(int)

df["INFECTED"] = df.INFECTED.astype(int)
df["TESTED"] = df.TESTED.astype(int)
```

```python
df.info()
```

```python
for i in range(0,len(df)):
    for j in df.columns:
        if df[j][i]==0:
            df[j][i]=np.nan        
```

```python
df
```

```python
df.info()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = (df.sort_values("INFECTED",ascending=False)["COUNTRY"].head(10)),
            y = (df.sort_values("INFECTED",ascending=False)["INFECTED"].head(10)),
            palette="CMRmap")

plt.figure(figsize=(15,7))
sns.barplot(x = (df.sort_values("TESTED",ascending=False)["COUNTRY"].head(10)),
            y = (df.sort_values("TESTED",ascending=False)["TESTED"].head(10)),
            palette="mako")

plt.figure(figsize=(15,7))
sns.barplot(x = (df.sort_values("RECOVERED",ascending=False)["COUNTRY"].head(10)),
            y = (df.sort_values("RECOVERED",ascending=False)["RECOVERED"].head(10)),
            palette="coolwarm")

plt.figure(figsize=(15,7))
sns.barplot(x = (df.sort_values("DECEASED",ascending=False)["COUNTRY"].head(10)),
            y = (df.sort_values("DECEASED",ascending=False)["DECEASED"].head(10)),
            palette="gist_rainbow_r")

plt.show()
```

```python
plt.pie((df.sort_values("INFECTED",ascending=False)["INFECTED"].head(10)),
        labels = (df.sort_values("INFECTED",ascending=False)["COUNTRY"].head(10)),
        colors = sns.color_palette("CMRmap"), 
        autopct = '%0.0f%%')
plt.title("INFECTED")
plt.show()

plt.pie((df.sort_values("TESTED",ascending=False)["TESTED"].head(10)),
        labels = (df.sort_values("TESTED",ascending=False)["COUNTRY"].head(10)),
        colors = sns.color_palette('mako'), 
        autopct = '%0.0f%%')
plt.title("TESTED")
plt.show()

plt.pie((df.sort_values("RECOVERED",ascending=False)["RECOVERED"].head(10)),
        labels = (df.sort_values("RECOVERED",ascending=False)["COUNTRY"].head(10)),
        colors = sns.color_palette('coolwarm'), 
        autopct = '%0.0f%%')
plt.title("RECOVERED")
plt.show()

plt.pie((df.sort_values("DECEASED",ascending=False)["DECEASED"].head(10)),
        labels = (df.sort_values("DECEASED",ascending=False)["COUNTRY"].head(10)),
        colors = sns.color_palette('gist_rainbow_r'), 
        autopct = '%0.0f%%')
plt.title("DECEASED")
plt.show()
```
